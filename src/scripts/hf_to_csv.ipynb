```python
import datasets
import pandas as pd
import typing
```

```python
output_dir = "../../data"

def load_dataset_pd(dataset_name: str) -> pd.DataFrame:
    dataset = datasets.load_dataset(dataset_name, split = "train")
    return pd.DataFrame(dataset)
```

```python

dataset_name = "preference-agents/enron-top-senders-train"
df = load_dataset_pd(dataset_name)
```

```python
# first 10 rows
df = df.head(10)
```

```python
# rename column infer_kaymann_rules_generated_email to kmann_train
df = df.rename(columns={"to_infer_kaymann_rules_generated_email": "kmann_train"})
# rename column naive_ftnk_generated_email to naive_train
df = df.rename(columns={"naive_ftnk_generated_email": "naive_train"})
```

```python
# save to csv
df.to_csv(f"{output_dir}/data.csv", index=False)
```
